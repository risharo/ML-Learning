```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv('Position_Salaries.csv')
df.head()
```

```python
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
X
```

## 1. Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

### Visulization LR

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## 2. Polynomial Regression (n = 2)

```python
poly2 = PolynomialFeatures(degree=2)
X_poly2 = poly2.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly2, y)
```

### Visualization of PR (N=2)

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr2.predict(X_poly2), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression deg = 2)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## 3. Polynomial Regression (n=3)

```python
poly3 = PolynomialFeatures(degree=3)
X_poly3 = poly3.fit_transform(X)
lr3 = LinearRegression()
lr3.fit(X_poly3, y)
```

### Visualization of PR (N=3)

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr3.predict(X_poly3), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression deg = 3)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## 3. Polynomial Regression (n=4)

```python
poly4 = PolynomialFeatures(degree=4)
X_poly4 = poly4.fit_transform(X)
lr4 = LinearRegression()
lr4.fit(X_poly4, y)
```

### Visualization of PR (N=4)

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr4.predict(X_poly4), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression deg = 4)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## 3. Polynomial Regression (n=5)

```python
poly5 = PolynomialFeatures(degree=5)
X_poly5 = poly5.fit_transform(X)
lr5 = LinearRegression()
lr5.fit(X_poly5, y)
```

### Visualization of PR (N=5)

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, lr5.predict(X_poly5), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression deg = 5)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

#### We will not choose n=5 polynomial regression due to overfitting. n=4 is best in this cases

### Smoothing of polynomial regression(deg =4) curve

```python
X_grid = np.arange(min(X), max(X), 0.1)

X_grid
```

```python
X_grid = X_grid.reshape(len(X_grid), 1)
X_grid
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X_grid, lr4.predict(poly4.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression deg = 4)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
