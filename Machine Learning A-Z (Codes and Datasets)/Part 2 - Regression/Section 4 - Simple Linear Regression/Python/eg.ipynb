```python
import random

random.seed(0)

data2 = []
col = ['Name of Company', 'Years of Working', 'Annual Revenue']

for i in range(1000):
    x = random.randint(1,8)
    years = random.randint(1,90)
    revenue = random.randint(5000000, 100000000)
    Alp = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

    data2.append([''.join(random.sample(Alp, x)), years, revenue])

df2 = pd.DataFrame(data2, columns= col)
df2
```
