```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
# Concatenate the data and target arrays
data_with_target = np.column_stack((iris.data, iris.target))

# Create the column names, combining feature names and the target name
columns = list(iris.feature_names) + ['target']

# Create a DataFrame
df = pd.DataFrame(data_with_target, columns=columns)

# Display the DataFrame
df.head()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state= 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=2)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn import tree

tree.plot_tree(model, filled= True)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_pred, y_test))
```

```python
print(classification_report(y_pred, y_test))
```

# Decision tree after pre-prunning

```python
## Preprunning
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(model, param_grid= parameter, cv=5, scoring= 'accuracy')
```

```python
cv.fit(X_train, y_train)
y_pred2 = cv.predict(X_test)
```

```python
y_pred2
```

```python
accuracy_score(y_pred2, y_test)
```
